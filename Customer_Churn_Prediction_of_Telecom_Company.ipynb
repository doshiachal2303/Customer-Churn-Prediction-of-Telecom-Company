#***Dividing Dependent And Independent Variables***
Before starting model building, it is important to divide the dataset into two parts: independent variables and dependent variable. The independent variables contain a list of those variables on which the concrete quality is dependent, while the dependent variable is the variable that is dependent on the values of other variables.

```python
# independent variable
x = df.drop(['Churn'], axis =1)
# dependent variable
y = df['Churn']
```

*   The program has a list of independent variables stored in x, while the dependent variable is in y.
*   The independent variables are Call Failure, Complaints, Subscription Length, Charge Amount, Seconds of Use, Frequency of Use, SMS Frequency, Distinct Called Numbers, Age Group, Tariff Plan, Status, Age, and Customer Value.
*   The dependent variable is Churn.





#***Splitting the Data***
To split the training and testing parts, we can use the scikit-learn module train_test_split

```python
# Splitting the Data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

#***Applying model***
Using a model in a project means putting the model's insights or predictions into practical use to achieve the project's objectives. This may include making real-time decisions, automating specific processes, improving efficiency, and ultimately delivering value to the organization or end-users. It is crucial to continually assess and update the applied model to ensure its accuracy and relevance in changing circumstances.


#***Decision Tree***

```python
# Applying Decision Tree
dtc = DecisionTreeClassifier(max_depth = 17) # max_depth values = [12, 15, 17]
dtc.fit(x_train, y_train)
y_pred = dtc.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy = round(accuracy*100, 2)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix: {cm}")
print(f"Classification report: {report}")
```

#***Plot of Actual Value vs Predicted Value for Decision Tree***

```python
# Plot of Actual Value vs Predicted Value for Decision Tree
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 42)
dtc = DecisionTreeClassifier(max_depth = 12)
dtc.fit(x_train, y_train)
y_pred = dtc.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize = (8, 6))
sb.heatmap(cm, annot = True, cmap = 'viridis', fmt = 'd', cbar = True)
plt.title('Plot of Actual Value vs Predicted Value for Decision Tree')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('Predicted', fontsize = 12)
plt.ylabel('Actual', fontsize = 12)
plt.show()
```

#***XG Boost***

```python
# Applying XG Boost
xgbc = XGBClassifier(max_depth = 12) # max_depths values = [8, 10, 12]
xgbc.fit(x_train, y_train)
y_pred = xgbc.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy = round(accuracy*100, 2)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix: {cm}")
print(f"Classification report: {report}")
```
